```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from textwrap import wrap
from IPython.display import display_html
import IPython
import re
# pd.set_option('display.max_colwidth', None)
# import seaborn as sns
# sns.set()
```

```python
df = pd.read_excel('respostas.xlsx', sheet_name='RESPOSTAS', skiprows=1)
# df.drop(['Carimbo de data/hora', 'Entrevistador:', 'Entrevistada(o):', 'Endereço', 'Casa:', 'Bairro/Distrito:'],axis=1, inplace=True)
df = df.applymap(lambda x: x.lower() if isinstance(x, str) else x)
df.replace('nao', 'não', inplace=True)
df.replace('nâo', 'não', inplace=True)
df = df.apply(lambda x: pd.to_numeric(x, errors='ignore'))
cols = list(df.columns)
```

```python
for i in range(1,223):
   html = '''
     <div style="text-align:center; width: 1000px">
        <h2>titulo</h2>
        <div style="position:relative; float:left; margin-right: 20px"><img src="imagem"/></div>
        <div style="position:relative; float:left;">data_frame</div>
     </div>
   '''
   c = cols[i]
   title = re.sub("\d+", "", c)
   title = re.sub("\.", "", title)
   title = re.sub("^\s", "", title)
   title = str(i) + '. ' + title
   dados = df[c].value_counts()
   t = sum(dados)
   dados = pd.DataFrame(data={'Contagem':dados, 'Porcentagem':np.round(100*dados/t,2)})
   dados_plot = dados.head()   
   dados_plot = dados_plot.sort_values(by=['Contagem'], ascending=True)

   data = dados_plot['Contagem']
   label = ['\n'.join(wrap(str(l), 40)) for l in dados_plot.index ]
   data.plot(kind='barh', figsize=(10,4))
   plt.xticks(fontsize=10)
   plt.box(on=None)
   plt.yticks(ticks= np.arange(len(data.index)), fontsize=10, labels=label)
   total = np.sum(data)
   for index, value in enumerate(data): 
       label = format(int(value), ',')
#        percent = '{:.0f}%'.format(value / total * 100)
#        plt.annotate(label + '\n' + percent, xy=(value+0.2, index-0.2), color='black', fontsize=12)
       plt.annotate(label, xy=(value+0.2, index-0.2), color='black', fontsize=10)
   plt.tight_layout()

   plt.savefig('img/'+str(i)+'.png')
   plt.close()
   
   html = '''
     <div style="text-align:center; position:relative; border: solid 1px; margin-bottom: 320px; border-radius: 5px; background: #f2f2f2">
        <h2>titulo</h2>
        <div style="position:relative; float:left; margin-right: 20px; margin-top: 1px;"><img src="imagem"/></div>
        <div style="position:relative; float:left; margin-top: 10px;"">data_frame</div>
     </div>
   '''
   html = html.replace('data_frame', dados.head(10).sort_values(by=['Contagem'], ascending=False).to_html()).replace('imagem', 'img/'+str(i)+'.png').replace('titulo', title)
   display_html(html, raw=True)
```
